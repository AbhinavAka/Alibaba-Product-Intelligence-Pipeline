```python
# Count Product Types
df["machine_type"].value_counts()
```

```python
# Visualize Product Types
df["machine_type"].value_counts().plot(kind="bar")

plt.title("Product Types")
plt.xlabel("Machine Type")
plt.ylabel("Count")

plt.show()
```

```python
# Check Supplier Types
df["supplier_type"].value_counts()
```

```python
# Visualize Supplier Types
df["supplier_type"].value_counts().plot(kind="pie")
plt.title("Supplier Types")
plt.show()
```

```python
# Check Supplier Locations
df["state"].value_counts()
```

```python
# Visualize Supplier Locations
df["state"].value_counts().plot(kind="bar")

plt.title("Suppliers by State")
plt.xlabel("State")
plt.ylabel("Number of Suppliers")

plt.show()
```

```python
# Check Product Prices
df["price"].head()
```

```python
# Cleaning the product price
df["price_clean"] = pd.to_numeric(df["price"].str.replace(",", "").str.extract(r'(\d+)')[0], errors="coerce")
```

```python
#Price Distribution chart
df["price_clean"].plot(kind="hist")
plt.title("Price Distribution")
plt.show()
```

```python
# Check Supplier Ratings
df["rating"].describe()
```

```python
# Check Years in Business
df["years_in_business"].describe()
```

```python
# Check duplicate data
df.duplicated().sum()
```

```python
# Drop duplicate
df = df.drop_duplicates()
```
